### Advanced 1:

- Unsupervised, as there is no delineated target in the data
- Clustering
- Centroid assigns one point as the center of the group, and see what falls within that range, and adjusts itself to include as many points as possible.
- Hierarchical begins as single groupings, and start to merge into larger groups, allowing you to decide at what point you want to pull the groupings.
- Density finds groups based on how near the data points are to eachother, making groups in any shape, and dropping outliers as noise.
- We would likely look to start with density, as buying habits will inherently have good data for clustering itself based on density.
- This data set could work for the client's problem. Before use, we may use mapping to convert many of the binary categorical data into numerical values for easier ML processing.
